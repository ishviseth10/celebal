## Step 1: Load Data
Import all required libraries and load the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set plot style
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)

# Load the dataset
df = pd.read_csv('../data/combined_dataset.csv')

print("✅ Dataset loaded successfully!")
print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")
print("\nColumn Names:")
print(df.columns.tolist())
```

## Step 2: Understand the Data
Explore data types, nulls, and summary statistics.

```python
# First few rows
print("=== First 5 Rows ===")
df.head()
```

```python
# Data types and null counts
print("=== Dataset Info ===")
df.info()
```

```python
# Summary statistics
print("=== Descriptive Statistics ===")
df.describe(include='all')
```

```python
# Missing values
print("=== Missing Values per Column ===")
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing %', ascending=False)
print(missing_df)
```

## Step 3: Data Cleaning
Fix data types, handle missing values, remove duplicates.

```python
# Keep only relevant columns for analysis
cols_to_keep = ['product_id', 'title', 'rating', 'ratings_count',
                'initial_price', 'discount', 'final_price', 'currency',
                'seller_name', 'category']
df = df[cols_to_keep].copy()
print("Kept relevant columns:", df.shape)
```

```python
# Clean price columns - remove currency symbols, quotes, commas
def clean_price(val):
    if pd.isnull(val):
        return np.nan
    val = str(val).replace('₹', '').replace(',', '').replace('"', '').strip()
    try:
        return float(val)
    except:
        return np.nan

df['final_price'] = df['final_price'].apply(clean_price)
df['initial_price'] = pd.to_numeric(df['initial_price'], errors='coerce')
df['discount'] = pd.to_numeric(df['discount'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['ratings_count'] = pd.to_numeric(df['ratings_count'], errors='coerce')

print("✅ Price columns cleaned.")
df[['initial_price', 'final_price', 'discount', 'rating', 'ratings_count']].describe()
```

```python
# Handle missing values
print("Missing values BEFORE cleaning:")
print(df.isnull().sum())

# Fill missing ratings with median
df['rating'].fillna(df['rating'].median(), inplace=True)
df['ratings_count'].fillna(0, inplace=True)

# Fill missing prices using the other column when possible
df['initial_price'].fillna(df['final_price'], inplace=True)
df['final_price'].fillna(df['initial_price'], inplace=True)

# Drop rows still missing price info
df.dropna(subset=['initial_price', 'final_price'], inplace=True)

print("\nMissing values AFTER cleaning:")
print(df.isnull().sum())
```

```python
# Remove duplicate records
before = len(df)
df.drop_duplicates(subset='product_id', inplace=True)
after = len(df)
print(f"✅ Removed {before - after} duplicate rows. Remaining: {after} rows.")
```

## Step 4: Feature Engineering
Create new meaningful columns from existing data.

```python
# 1. Price Difference Column (savings amount)
df['price_difference'] = df['initial_price'] - df['final_price']
df['price_difference'] = df['price_difference'].clip(lower=0)  # no negatives

# 2. Popularity Metric = rating * log(ratings_count + 1)
df['popularity_score'] = df['rating'] * np.log1p(df['ratings_count'])

print("✅ New columns created: 'price_difference' and 'popularity_score'")
df[['title', 'initial_price', 'final_price', 'price_difference', 'rating', 'ratings_count', 'popularity_score']].head(10)
```

## Step 5: Analysis

### 5a. Univariate Analysis
Analyzing individual columns one at a time.

```python
# Distribution of ratings
print("Rating value counts (rounded):")
print(df['rating'].round(1).value_counts().sort_index())
```

```python
# Category distribution
print("Products per Category:")
print(df['category'].value_counts())
```

### 5b. Bivariate Analysis
Relationship between two variables.

```python
# Correlation between rating and ratings_count
corr = df[['rating', 'ratings_count', 'price_difference', 'popularity_score', 'final_price']].corr()
print("Correlation Matrix:")
print(corr.round(2))
```

### 5c. Category-Level Analysis

```python
# Average price and rating by category
cat_analysis = df.groupby('category').agg(
    avg_final_price=('final_price', 'mean'),
    avg_rating=('rating', 'mean'),
    avg_discount=('discount', 'mean'),
    product_count=('product_id', 'count')
).round(2).sort_values('product_count', ascending=False)

print("Category-Level Summary:")
print(cat_analysis)
```

## Step 6: Visualizations

```python
# 1. Histogram of Final Price
plt.figure(figsize=(10, 5))
df['final_price'].clip(upper=df['final_price'].quantile(0.95)).hist(bins=40, color='steelblue', edgecolor='white')
plt.title('Distribution of Final Price (up to 95th percentile)', fontsize=14)
plt.xlabel('Final Price (₹)')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.savefig('plot_price_distribution.png', dpi=100)
plt.show()
print("✅ Plot saved.")
```

```python
# 2. Bar Chart - Number of Products per Category
plt.figure(figsize=(14, 6))
cat_counts = df['category'].value_counts().head(15)
sns.barplot(x=cat_counts.values, y=cat_counts.index, palette='viridis')
plt.title('Top 15 Categories by Product Count', fontsize=14)
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.tight_layout()
plt.savefig('plot_category_count.png', dpi=100)
plt.show()
```

```python
# 3. Boxplot - Rating by Category (top 8 categories)
top_cats = df['category'].value_counts().head(8).index
plt.figure(figsize=(14, 6))
sns.boxplot(data=df[df['category'].isin(top_cats)], x='category', y='rating', palette='Set2')
plt.title('Rating Distribution across Top 8 Categories', fontsize=14)
plt.xlabel('Category')
plt.ylabel('Rating')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig('plot_rating_boxplot.png', dpi=100)
plt.show()
```

```python
# 4. Scatter: Popularity Score vs Final Price
plt.figure(figsize=(10, 5))
plt.scatter(df['final_price'].clip(upper=5000), df['popularity_score'], alpha=0.4, color='coral', s=15)
plt.title('Popularity Score vs Final Price', fontsize=14)
plt.xlabel('Final Price (₹)')
plt.ylabel('Popularity Score')
plt.tight_layout()
plt.savefig('plot_popularity_vs_price.png', dpi=100)
plt.show()
```

```python
# 5. Heatmap of correlations
plt.figure(figsize=(8, 5))
corr_cols = ['rating', 'ratings_count', 'final_price', 'price_difference', 'popularity_score', 'discount']
sns.heatmap(df[corr_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=14)
plt.tight_layout()
plt.savefig('plot_heatmap.png', dpi=100)
plt.show()
```

## Step 7: Key Insights & Business Implications

### 📌 Key Findings

1. **Price Distribution**: Most products are priced in the affordable-to-mid range. A small number of premium products skew the distribution upward.

2. **Category Breakdown**: Categories like backpacks and clothing dominate the dataset in product count, suggesting these are high-inventory segments.

3. **Ratings**: The majority of products have ratings between 3.5 and 4.5, indicating generally positive customer satisfaction. Very few products fall below 3.

4. **Discounts**: Products with higher initial prices tend to offer larger absolute discounts, making them appear more attractive to deal-seeking customers.

5. **Popularity Score**: Products with more reviews don't always have the highest ratings — high volume + moderate rating can still outperform a 5-star product with only a few reviews.

### 💼 Business Implications

- **Stock high-rated, high-popularity products** to maximize customer satisfaction and repeat purchases.
- **Review low-rated categories** for quality issues or misleading product descriptions.
- **Pricing strategy**: Mid-range products see higher engagement — deep discounts on premium items can drive conversion.
- **Focus marketing on top categories** since they have the most customer interest and review volume.
